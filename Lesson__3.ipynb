```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HumayDS/Dataset3/refs/heads/main/customer_value_analysis.csv')
df.head()
```

# Sual 1

```python
#1.1
df.isnull().sum()
df = df.dropna()
```

```python
# 1.2
# Convert date columns to datetime format
df["first_purchase_date"] = pd.to_datetime(df["first_purchase_date"])  # convert to datetime
df["last_purchase_date"] = pd.to_datetime(df["last_purchase_date"])
# Create customer_lifetime (difference in days)
df["customer_lifetime"] = (df["last_purchase_date"] - df["first_purchase_date"]).dt.days  # calculate days difference

# Create purchase_frequency
df["purchase_frequency"] = df["total_orders"] / df["customer_lifetime"]  # orders per day
```

```python
#1.3
df.describe().T
```

# Sual 2

```python
df.head()
```

```python
2.1 # Calculate CLV
df["CLV"] = df["avg_order_value"] * df["purchase_frequency"] * df["customer_lifetime"]
```

```python
#2.2
# Segment customers into 3 groups using qcut
df["CLV_segment"] = pd.qcut(
    df["CLV"],
    q=3,  # divide into 3 equal groups
    labels=["low-value", "medium-value", "high-value"])
```

#Sual 3

```python
# 3.1
df["CLV"].hist()
```

```python
#3.2

segment_analysis = df.groupby('CLV_segment').agg({
    'CLV': 'mean',
    'total_orders': 'mean',
    'total_spent': 'mean',
    'customer_lifetime': 'mean',
    'purchase_frequency': 'mean'
}).round(2)

segment_analysis
```
