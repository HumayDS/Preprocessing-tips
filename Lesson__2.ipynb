```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HumayDS/Dataset3/refs/heads/main/digital_marketing_performance.csv')
df.head()
```

```python
#1.1
# Check for duplicate rows
duplicates = df.duplicated()
print("Duplicate rows count:", duplicates.sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify removal
print("After removal:", df.duplicated().sum())
```

```python
#1.2
# Create CTR (Click-Through Rate) column
df['CTR'] = df['clicks'] / df['impressions']
```

```python
# # Select numerical columns
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Display correlation matrix
print(corr_matrix)
```

```python
  #2.2
# Group by campaign and calculate totals
campaign_summary = df.groupby('campaign_id').agg(
    total_clicks=('clicks', 'sum'),
    total_conversions=('conversions', 'sum')
).reset_index()

# Calculate conversion rate (correct way)
campaign_summary['conversion_rate'] = (
    campaign_summary['total_conversions'] / campaign_summary['total_clicks']
)

# Find best and worst campaigns
best_campaign = campaign_summary.loc[campaign_summary['conversion_rate'].idxmax()]
worst_campaign = campaign_summary.loc[campaign_summary['conversion_rate'].idxmin()]

print("Best campaign:")
print(best_campaign)

print("\nWorst campaign:")
print(worst_campaign)
```
