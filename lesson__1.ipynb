```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HumayDS/Dataset3/refs/heads/main/customer_transactions_messy.csv')
```

```python
df.head(25)
```

#Sual 1

```python
#1.1
df.isnull().sum()
# Calculate median values for numeric columns
quantity_median = df['quantity'].median()
price_median = df['price'].median()

# Fill missing values with median
df['quantity'] = df['quantity'].fillna(quantity_median)
df['price'] = df['price'].fillna(price_median)
```

```python
# Convert the date column to datetime format
# errors='coerce' will convert invalid formats to NaT
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
#1.2
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
#1.3
# Create a new column for total spending per transaction
df['total_spent'] = df['quantity'] * df['price']
```

# Sual 2

```python
#2.1
# Group by product and sum total quantity sold
top_products = df.groupby('product')['quantity'].sum().sort_values(ascending=False).head(5)

print(top_products)
```

```python
#2.2
# Extract month from date
df['month'] = df['date'].dt.to_period('M')

# Aggregate total sales per month
monthly_trend = df.groupby('month')['total_spent'].sum()

print(monthly_trend)
```

```python
#2.3
# Calculate IQR (Interquartile Range)
Q1 = df['total_spent'].quantile(0.25)
Q3 = df['total_spent'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[(df['total_spent'] < Q1 - 1.5 * IQR) |
              (df['total_spent'] > Q3 + 1.5 * IQR)]

print(outliers)
```

```python
# Count number of transactions per customer
customer_orders = df.groupby('customer_id')['transaction_id'].count()

# Identify one-time vs repeat customers
one_time_customers = (customer_orders == 1).sum()
repeat_customers = (customer_orders > 1).sum()

print("One-time customers:", one_time_customers)
print("Repeat customers:", repeat_customers)
```
